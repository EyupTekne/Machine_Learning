```python
import pandas as pd
import re

# The file name here must be the same as the file name you uploaded.
# If it is different, change the “g-code square.txt” part.
with open("g-code square.txt", "r") as f:
    lines = f.readlines()

pattern = re.compile(r"N(\d+)\s+([A-Z0-9]+)?(?:\s+X([-0-9.]+))?(?:\s+Y([-0-9.]+))?(?:\s+Z([-0-9.]+))?(?:\s+F([-0-9.]+))?")

data = []

for line in lines:
    m = pattern.search(line)
    if m:
        data.append([
            int(m.group(1)),
            m.group(2),
            float(m.group(3)) if m.group(3) else None,
            float(m.group(4)) if m.group(4) else None,
            float(m.group(5)) if m.group(5) else None,
            float(m.group(6)) if m.group(6) else None,
        ])

df = pd.DataFrame(data, columns=["Line", "GCodeType", "X", "Y", "Z", "FeedRate"])
df
```

```python
# Basic Information
print("==== df.info() ====")
print(df.info())

print("\n==== df.describe() ====")
print(df.describe())  # Summary statistics for numerical columns

print("\n==== GCodeType dağılımı ====")
print(df["GCodeType"].value_counts())
```

```python
import matplotlib.pyplot as plt

# X-Y Take the points, but only the valid ones
points = df.dropna(subset=["X", "Y"])[["X", "Y"]].values

# Reorder the points according to the CNC drawing sequence:
# 1) (0,0)
# 2) (50,0)
# 3) (50,50)
# 4) (0,50)
# 5) (0,0)
ordered = sorted(points, key=lambda p: (p[0], p[1]))

# The sequence could be: (0,0), (0,50), (50,0), (50,50)
# But the actual CNC drawing is right → up → left → down.
# So we will manually specify the correct sequence.:

ordered_points = [
    (0,0),
    (50,0),
    (50,50),
    (0,50),
    (0,0)
]

xs = [p[0] for p in ordered_points]
ys = [p[1] for p in ordered_points]

plt.figure(figsize=(6,6))
plt.plot(xs, ys, marker='o', linewidth=3)
plt.title("CNC Toolpath (X-Y) - Correct Closed Square")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()



```
